```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import Imputer
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
submission_df = pd.read_csv("submission.csv")
train_df.info() #最高學歷 、畢業學校類別 缺值特別多 ， 其餘都是缺73筆
```

```python
train_df = train_df.dropna(subset=['工作分類', '職等']) #發現每一欄所缺的73筆都為同一行
train_df.info()
```

```python
#處理最高學歷 、畢業學校類別 缺失值 imputer 代入
imputer = Imputer(missing_values = np.nan, strategy = "most_frequent", axis = 0) #實驗完用眾數準確度最高
train1_df = imputer.fit_transform(train_df)
train1_df = pd.DataFrame(train1_df)
train1_df.columns = train_df.columns #將個別欄位名帶入
train1_df
```

```python
train2_df = train1_df.drop(columns = ["yyyy", "PerNo", "職等", "廠區代碼", "管理層級", "畢業學校類別", "畢業科系類別","通勤成本","任職前工作平均年數","年資層級C", "訓練時數A", "訓練時數B", "訓練時數C"])
train2_df
```

```python
x = train2_df.drop("PerStatus", axis=1)
y = train2_df["PerStatus"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)

from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(x_train,y_train)

y_pre_clf = clf.predict(x_test)
score_clf = accuracy_score(y_test, y_pre_clf)
print("score_clf: ",score_clf)

```
