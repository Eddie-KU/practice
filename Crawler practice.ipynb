```python
#ppt整頁內文爬取( 給分數、 關鍵字提取 )

# requests 套件 : 有內含header
# jieba: 中文斷詞套件
from jieba.analyse import extract_tags #提取關鍵字
import requests
from bs4 import BeautifulSoup

url = "https://www.ptt.cc/bbs/biker/M.1602870115.A.4AB.html"

response = requests.get(url)
html = BeautifulSoup(response.text)

#尋找標籤元素為div , 屬性id = maim-content 符合的第一個值
content = html.find("div", id = "main-content")

#尋找標籤元素為 span , 屬性class = article-meta-value 的所有值 (串列)
meta = content.find_all("span", class_ = "article-meta-value" )

print("ID: ",meta[0].text) #串列裡第一個位置
print("看板: ",meta[1].text)
print("標題: ",meta[2].text)
print("時間: ",meta[3].text)

#extract 丟掉元素
metas = content.find_all("div", class_ = "article-metaline")
for m in metas:
    m.extract()
ms = content.find_all("div", class_ = "article-metaline-right")
for m in metas:
    m.extract()

pushes = content.find_all("div", class_ = "push")  
score = 0
push = []
for p in pushes:
    tag = p.find("span", class_ = "push-tag").text  #推、噓、-> 都有push-tag
    if "推" in tag:
        score += 2
    elif"噓"in tag:
        score -= 2
    p.extract  #不需要 p ，丟掉
print("分數:", score)
print("本文:", content.text)
print("關鍵字", extract_tags(content.text, 10)) #提取本文關鍵字
```

```python
#爬取需突破18禁限制的八卦版標題文字

# requests套件 :可以內含header(信件)自動去突破80%的網站，但遇到18進限制需要設定coockies突破
import requests
#BeautifulSoup:用來解析 html的工具
import bs4

URL = "https://www.ptt.cc/bbs/Gossiping/index.html"

#設定header與cookie
my_headers = {'cookie':'over18=1;'}


#發送請求到八卦版
response = requests.get(URL, headers = my_headers)


#將得到的回應給 bs4 分析 
content = bs4.BeautifulSoup(response.text,"html.parser")



#在元素裡找出所有標籤名為"div" class為 "title"
titles = content.find_all("div","title")


#萃取文字，只保留文字
#用for 迴圈將每個爬到的 title放到 List裡 
for t in titles:
    #strip()去處空白
    print(t.text.strip())
```

```python
#一次爬ptt多頁標題(累加)

import requests
import bs4
import time

#定義一個獲取單一網頁html內容並返回的函式
def one_page(URL):
    
    my_headers = {'cookie': 'over18=1;'}
    response = requests.get(URL, headers = my_headers)
    content = bs4.BeautifulSoup(response.text,"html.parser")
    titles = content.find_all('div','title')
       
    for t in titles:
        print(t.text.strip())  
        

start = 38969 # 設定開始爬取的網頁 
pages = 5    # 設定要從開始頁面往後爬幾頁
end = start - pages



#使用for in range 做多頁爬蟲
for i in range(start,end,-1):
    url = "https://www.ptt.cc/bbs/Gossiping/index"+str(i)+".html"
    
    one_page(url)   # 執行單頁爬蟲
    time.sleep(1)    # 等待一秒再繼續迴圈 避免被 PTT封鎖請求
```

```python
#爬取日本餐廳排名網站 多頁資訊爬取存成csv檔

from urllib.request import urlopen
from urllib.error import HTTPError #處理error
from bs4 import BeautifulSoup
import pandas as pd

#創建一個表格
df = pd.DataFrame(columns=["Total", "JA", "EN", "url"])
page =  51


while True:
    url = "https://tabelog.com/tw/tokyo/rstLst/"+str(page)+"/?SrtT=rt"
    print("處理url", url)

    
    try:
        response = urlopen(url)
    except HTTPError:
        print("最後一頁了")
        break 
    html = BeautifulSoup(response) #產生一堆元素 


    
    for r in html.find_all('li', class_="list-rst"):

        
        ja = r.find("small", class_="list-rst__name-ja") #尋找標籤元素為small , 屬性class =list-rst__name-ja 符合的第一個值
        en = r.find("a", class_ = "list-rst__name-main") #找標籤元素為a , 屬性class =list-rst__name 符合的第一個值
        ratings = r.find_all("b",class_ = "c-rating__val")#尋找標籤元素為 b , 屬性class = c-rating__val 的所有值 
        
        
        #萃取(.text) 
        print(ratings[0].text, ja.text, en.text, en["href"])
        
       
        #使用pd.series創建
        s = pd.Series([ratings[0].text,ja.text, en.text, en["href"]], index = ["Total", "JA", "EN", "url"])
        
        
        # 將所創建的series 放進來 df 要忽略原本index 
        df = df.append(s, ignore_index = True)
    
    
    #增加page
    page += 1

    
#儲存成csv檔
df.to_csv("tabelog.csv", encoding = "utf-8", index = False)
```
