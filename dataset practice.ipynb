```python
#篩選,  概念 : df[[跟列一樣多的 boolean]] 要留的 True   不留的 False
df.head(5)[[True, False, True, False, True]] #只會有第1, 3, 5列


#篩選, 自訂義條件函式操作 Ex:
def fflow(n):
    if n > 2000 and n <  5000:
        return True
    else:
        return False

f = df["comments"].apply(fflow)
df[f]
```

```python
#定義一個找尋所有出現 computer 的方法
def fflow(s):
    
    #eval() 把傳入的字串當成 py程式在執行 eval("print(3+2)") # a = [1, 2, 3]
    taglist = eval(s)     
    
    if "computers" in taglist:
        return True
    
    else:
        return False
       
df[ df["tags"].apply(fflow)]
```
